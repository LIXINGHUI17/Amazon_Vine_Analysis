```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change the path if needed.)
vine_table_load = "vine_table.csv"

# Read the School Data and Student Data and store into a Pandas DataFrame
df = pd.read_csv(vine_table_load)
```

```python
df.head(20)
```

```python
# Filter the data and create a new DataFrame or table to retrieve all the rows where 
# the total_votes count is equal to or greater than 20 to pick reviews that are more 
# likely to be helpful and to avoid having division by zero errors later on.
df1 = df[(df.total_votes >= 20)]
df1.head(20)
```

```python
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame 
# or table to retrieve all the rows where the number of helpful_votes divided by 
# total_votes is equal to or greater than 50%.
df2 = df1[df1.helpful_votes/df1.total_votes >=0.5 ]
df2.head(20)
```

```python
# Filter the DataFrame or table created in Step 2, and create a new DataFrame 
# or table that retrieves all the rows where a review was written as part of the 
# Vine program (paid), vine == 'Y'.
df3 = df2[(df2.vine == 'Y')]
df3.head(20)
```

```python
df4 = df2[(df2.vine == 'N')]
df4.head(20)
```

```python
# Total number of review
total = df.review_id.count()
total
```

```python
# the number of 5-star reviews
five_stars = df.review_id[df.star_rating == 5].count()
five_stars
```

```python
# the percentage of 5-star reviews for paid
five_stars_paid_review = df3.review_id[df3.star_rating == 5].count()
five_stars_paid_percentage = (five_stars_paid_review / total) * 100
five_stars_paid_percentage
```

```python
five_stars_paid_review
```

```python
# the percentage of 5-star reviews for unpaid
five_stars_unpaid_review = df4.review_id[df4.star_rating == 5].count()
five_stars_unpaid_percentage = (five_stars_unpaid_review / total) * 100
five_stars_unpaid_percentage
```
